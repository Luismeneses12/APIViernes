algoritmo para el cálculo de áreas y volúmenes.

```python
import math

elegirUnaOpcion = int(input(""" esta es una calculadora de  para medir en area y el volumen
                              #############################################################
                              en las diferentes opciones puede encontras la soluciona sacar el areas.
                                               opcion
                                               1.cuadrado
                                               2.rectangulo
                                               3.triangulo
                                               4.circulo
                              ____________________________________________________
                                          valumen de cuerpo geometricos
                                              5.cubo
                                              6.ortoedro
                                              7.cilindro
                                              8.esfera
                                              9.cono
                                              10.piramide
                                                              """))
match elegirUnaOpcion:
  case 1:
    lado = float(input("digite un lado: "))
    operacion = lado * lado
    print(f"El area del cuadrado es {operacion}")
  case 2:
    base = float(input("digite la base: "))
    altura = float(input("digite la altura: "))
    operacion = base * altura
    print(f"El area del rectangulo es {operacion}")
  case 3:
    base = float(input("digite la base: "))
    altura = float(input("digite la altura: "))
    operacion = (base * altura) / 2
    print(f"El area del triangulo es {operacion}")
  case 4:
    radio = float(input("digite el radio: "))
    operacion = math.pi * (radio ** 2)
    print(f"El area del circulo es {operacion}")
  case 5:
    lado = float(input("digite un lado: "))
    operacion = lado ** 3
    print(f"El volumen del cubo es {operacion}")
  case 6:
    largo = float(input("digite el largo: "))
    ancho = float(input("digite el ancho: "))
    altura = float(input("digite la altura: "))
    operacion = largo * ancho * altura
    print(f"El volumen del ortoedro es {operacion}")
  case 7:
    radio = float(input("digite el radio de la base: "))
    altura = float(input("digite la altura: "))
    operacion = math.pi * (radio ** 2) * altura
    print(f"El volumen del cilindro es {operacion}")
  case 8:
    radio = float(input("digite el radio: "))
    operacion = (4/3) * math.pi * (radio ** 3)
    print(f"El volumen de la esfera es {operacion}")
  case 9:
    radio = float(input("digite el radio de la base: "))
    altura = float(input("digite la altura: "))
    operacion = (1/3) * math.pi * (radio ** 2) * altura
    print(f"El volumen del cono es {operacion}")
  case 10:
    area_base = float(input("digite el area de la base: "))
    altura = float(input("digite la altura: "))
    operacion = (1/3) * area_base * altura
    print(f"El volumen de la piramide es {operacion}")
  case _:
    print("Opcion no valida. Por favor, elija un numero entre 1 y 10.")
```
